```python
# ============================================================
# ğŸ¤– CodTech AI Internship - Task 4: GPT-2 Text Generator
# ğŸ“Œ This script generates coherent paragraphs using GPT-2
#     based on user prompts. It uses HuggingFace Transformers.
# Developed by: [Your Name]
# ============================================================

from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
import os
import datetime
import textwrap

def load_model():
    """
    Loads the GPT-2 model and tokenizer from Hugging Face Hub.
    Returns both for use in text generation.
    """
    print("ğŸ“¦ Loading GPT-2 model and tokenizer...")
    tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
    model = GPT2LMHeadModel.from_pretrained("gpt2")
    model.eval()  # Set model to inference mode
    return tokenizer, model

def generate_text(prompt, tokenizer, model, max_length=100, top_k=50):
    """
    Generates text from a prompt using GPT-2.
    Parameters:
        - prompt: User input text to continue
        - max_length: Maximum length of output (tokens)
        - top_k: Sampling diversity (lower = more focused)
    Returns:
        - A generated string
    """
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate continuation
    outputs = model.generate(
        inputs,
        max_length=max_length,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        top_k=top_k,
        pad_token_id=tokenizer.eos_token_id
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)

def save_to_file(text, prompt_snippet):
    """
    Saves the generated text to a timestamped file.
    """
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"generated_{prompt_snippet}_{timestamp}.txt".replace(" ", "_")
    with open(filename, "w", encoding="utf-8") as f:
        f.write(text)
    print(f"âœ… Text saved as '{filename}'")

def display_wrapped_text(text, width=90):
    """
    Prints the generated text wrapped to specified width.
    Improves readability in the terminal.
    """
    print("ğŸ“ Generated Text:\n" + "=" * 50)
    wrapped = textwrap.fill(text, width=width)
    print(wrapped)
    print("=" * 50)

def main():
    """
    Main function to run the text generation tool.
    Accepts user prompt, generates text, and optionally saves it.
    """
    print("ğŸ¤– TEXT GENERATION - CodTech Task 4")
    print("=====================================\n")

    # Get user input
    prompt = input("âœï¸ Enter a prompt (e.g., 'The future of AI is'): ").strip()
    if not prompt:
        prompt = "Technology is evolving at a rapid pace"
        print(f"âš ï¸ Empty input detected. Using default prompt: {prompt}")

    # Optional: set maximum token length
    max_len = input("ğŸ“ Max length (default 100): ").strip()
    max_len = int(max_len) if max_len.isdigit() else 100

    # Load model and tokenizer
    tokenizer, model = load_model()

    print("\nğŸ”® Generating text...\n")
    generated = generate_text(prompt, tokenizer, model, max_length=max_len)

    # Display formatted output
    display_wrapped_text(generated)

    # Save option
    save = input("\nğŸ’¾ Do you want to save the output? (y/n): ").strip().lower()
    if save == "y":
        save_to_file(generated, prompt[:10])

if __name__ == "__main__":
    main()
```
